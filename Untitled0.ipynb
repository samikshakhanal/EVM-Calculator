```python
# === EARNED VALUE MANAGEMENT CALCULATOR ===


def run_evm():
    try:
        PV = float(input("Enter Planned Value (PV): "))
        EV = float(input("Enter Earned Value (EV): "))
        AC = float(input("Enter Actual Cost (AC): "))
        BAC = float(input("Enter Budget at Completion (BAC): "))
    except ValueError:
        print("Error!!! Please enter numbers only.")
        return

    # Calculations
    CPI = EV / AC
    SPI = EV / PV
    CV = EV - AC
    SV = EV - PV
    EAC = BAC / CPI

    # Results
    print("\n=== EVM RESULTS ===")
    print(f"Cost Performance Index (CPI): {CPI:.3f}")
    print(f"Schedule Performance Index (SPI): {SPI:.3f}")
    print(f"Cost Variance (CV): {CV:.2f}")
    print(f"Schedule Variance (SV): {SV:.2f}")
    print(f"Estimate at Completion (EAC): {EAC:.2f}")

# Run the calculator
run_evm()
```

# Earned Value Management (EVM) Calculator
This is a simple Python-based calculator for basic Earned Value Management calculations, including parameters such as:

- Cost Performance Index (CPI)
- Schedule Performance Index (SPI)
- Cost Variance (CV)
- Schedule Variance (SV)
- Estimate at Completion (EAC)

### Why this project?
EVM is a core part of Construction Management, Project monitoring, Project Scheduling, and Project control.  
This calculator helps understand cost and schedule performance using fundamental formulas.

### Defining Parameters used in EVM Calculator
Parameters used in this simple calculator are:
- Planned Value: Expression of work planned to complete by certain date in terms of monetary units
- Earned Value: Expression of work that has actually been completed in terms of monetary units
- Actual Cost: Amount of money that has been used to complete the work so far
- Budget at Completion: Total Planned budget for the entire project

PV = 5000
EV = 4500
AC = 4800
